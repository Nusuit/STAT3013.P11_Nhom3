```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV files
order_items = pd.read_csv('/content/drive/MyDrive/Phân tích thống kê/Project/olist_dataset/olist_order_items_dataset.csv')
order_reviews = pd.read_csv('/content/drive/MyDrive/Phân tích thống kê/Project/olist_dataset/olist_order_reviews_dataset.csv')
products = pd.read_csv('/content/drive/MyDrive/Phân tích thống kê/Project/olist_dataset/olist_products_dataset.csv')

# Merge the order_items and order_reviews datasets based on 'order_id'
merged_data = pd.merge(order_items, order_reviews, on='order_id', how='inner')

# Merge the merged_data with products dataset based on 'product_id'
full_data = pd.merge(merged_data, products, on='product_id', how='inner')

# Identify the lowest reviews (e.g., review_score <= 2)
low_reviews = full_data[full_data['review_score'] <= 2]

# Count the number of negative reviews by product category
category_negative_counts = low_reviews['product_category_name'].value_counts()

# Find the category with the highest number of negative reviews
most_negative_category = category_negative_counts.idxmax()

# Count the number of negative reviews by product in the most negative category
negative_products_in_category = low_reviews[low_reviews['product_category_name'] == most_negative_category]['product_id'].value_counts()

# Get the top 10 products (product_id) with the most negative reviews
top_10_negative_products = negative_products_in_category.head(10).reset_index()
top_10_negative_products.columns = ['product_id', 'negative_count']
```

```python
# Get the top 10 product categories with the most negative reviews
top_10_negative_categories = category_negative_counts.head(10)

# Visualize the top 10 product categories with the most negative reviews
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_negative_categories.values, y=top_10_negative_categories.index, palette="Blues_r")
plt.title("Top 10 Product Categories with the Most Negative Reviews")
plt.xlabel("Number of Negative Reviews")
plt.ylabel("Product Categories")
plt.show()
```

```python
# Visualize the top 10 products with the most negative reviews
plt.figure(figsize=(12, 6))
sns.barplot(
    x='negative_count',
    y='product_id',
    data=top_10_negative_products,
    palette="Blues_r"
)
plt.title(f"Top 10 Products (Product ID) in the Category '{most_negative_category}' with the Most Negative Reviews")
plt.xlabel("Number of Negative Reviews")
plt.ylabel("Product ID")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
import nltk

# Step 1: Rename columns for consistency
full_data = full_data.rename(columns={'review_comment_message': 'text', 'review_score': 'label'})

# Step 4: Filter necessary columns
full_data = full_data[['text', 'label', 'product_category_name']].dropna()

# Step 5: Convert review scores to binary labels (0 for negative, 1 for positive)
full_data['label'] = full_data['label'].apply(lambda x: 0 if x <= 2 else 1)

# Step 6: Select the top 5 product categories with the most reviews
top_categories = full_data['product_category_name'].value_counts().head(5).index

# Step 7: Filter the data to include reviews from only these top 5 categories
filtered_data = full_data[full_data['product_category_name'].isin(top_categories)]

# Split the data into negative and positive reviews
negative_reviews = filtered_data[filtered_data['label'] == 0]
positive_reviews = filtered_data[filtered_data['label'] == 1]

# Step 8: Download stopwords
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('portuguese')

# Step 9: Define function to analyze most common words in each product category
def analyze_top_words(reviews, sentiment):
    category_words = {}
    for category in reviews['product_category_name'].unique():
        category_reviews = reviews[reviews['product_category_name'] == category]

        # Use TF-IDF to extract the most common words
        tfidf = TfidfVectorizer(stop_words=stop_words, max_features=10)  # Top 10 words
        tfidf_matrix = tfidf.fit_transform(category_reviews['text'].fillna(''))

        # Calculate TF-IDF scores
        word_scores = pd.DataFrame(tfidf_matrix.sum(axis=0), columns=tfidf.get_feature_names_out()).T
        word_scores.columns = ['score']
        word_scores = word_scores.sort_values(by='score', ascending=False)

        # Store the top 10 words for each category
        category_words[category] = word_scores.head(10).index.tolist()

    return category_words

# Analyze key words for negative and positive reviews
negative_words = analyze_top_words(negative_reviews, 'negative')
positive_words = analyze_top_words(positive_reviews, 'positive')

# Step 10: Define function to visualize top words
def visualize_top_words(word_dict, sentiment):
    for category, words in word_dict.items():
        plt.figure(figsize=(8, 5))
        sns.barplot(x=[1]*len(words), y=words, palette="Blues" if sentiment == 'negative' else "Greens")
        plt.title(f"Top words in {sentiment} reviews for category '{category}'")
        plt.xlabel("Frequency (TF-IDF)")
        plt.ylabel("Words")
        plt.show()

# Visualize common words in negative reviews for the top 5 categories
visualize_top_words(negative_words, 'negative')

# Visualize common words in positive reviews for the top 5 categories
visualize_top_words(positive_words, 'positive')
```
